```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(12, 8))

# Dessin des nœuds
circles = [(0.1, 0.9, 'Fièvre?'),
           (0.4, 0.7, 'Toux/Mal de gorge?'),
           (0.4, 0.5, 'Courbatures?'),
           (0.4, 0.3, 'Exposition à la grippe?'),
           (0.8, 0.7, 'Symptômes gastro?'),
           (0.8, 0.5, 'Fatigue extrême?'),
           (0.8, 0.3, 'Vaccination?'),
           (0.1, 0.5, 'Durée > 3 jours?'),
           (0.4, 0.1, 'Essoufflement?'),
           (0.8, 0.1, 'Maux de tête?')]

for x, y, label in circles:
    circle = mpatches.Circle((x, y), 0.1, ec="none", color='skyblue')
    ax.add_patch(circle)
    ax.text(x, y, label, ha="center", va="center", fontsize=10)

# Dessin des branches (lignes)
lines = [(0.1, 0.8, 0.4, 0.7),  # Ligne entre "Fièvre?" et "Toux/Mal de gorge?"
         (0.1, 0.8, 0.8, 0.7),  # Ligne entre "Fièvre?" et "Symptômes gastro?"
         # (Ajouter plus de lignes ici pour connecter tous les nœuds)
         ]

for x1, y1, x2, y2 in lines:
    ax.plot([x1, x2], [y1, y2], 'k-')

ax.axis('off')
plt.show()
```
